```
import pandas as pd
import numpy as np
```

```
names = ['student_id','internships','training']
df = pd.read_csv('drive/My Drive/Anjali.csv',names=names)
```

```
df.shape
```

```
df1 = pd.read_csv('drive/My Drive/profiles.csv',names = 'p')
df1
```

```
l = df1['p'].values.tolist()
```

```
def convert(s):
    if s in l:
        return s
    else:
        df.fillna
    
df.internships = df.internships.map(convert)
df.head(10)
type(df)
```

```
df = df.dropna()
df = df.drop_duplicates()
df.shape
```

```
df=df.groupby(['student_id','training'])['internships'].apply(lambda x: ','.join(x.astype(str))).reset_index()
#df=df.groupby(['student_id','internships'])['training'].apply(lambda x: ','.join(x.astype(str))).reset_index()
df
```

```
split = int(0.8 * df.shape[0])
train_data = df[:split]
test_data = df[split:]
```

```
df = df.dropna()
df.head(-1)
train_data = train_data.dropna()
test_data = test_data.dropna()
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(df['internships'])
from sklearn.metrics.pairwise import cosine_similarity,linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix1 = tf.fit_transform(train_data['internships'])
from sklearn.metrics.pairwise import cosine_similarity,linear_kernel
cosine_sim = linear_kernel(tfidf_matrix1, tfidf_matrix1)
```

```
df = df.reset_index()
titles = df['training']
indices = pd.Series(df.index, index= df['training'])
print(titles)
```

```
train_data = train_data.reset_index()
titles1 = train_data['training']
indices1 = pd.Series(train_data.index, index= train_data['training'])
print(titles1)
```

```
def get_recommendations(title):
    print(type(title))
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    training_indices = [i[0] for i in sim_scores]
    return titles.iloc[training_indices]
```
